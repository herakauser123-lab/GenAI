#Task 1: Replace Hardcoded Prompt


In traditional Python code, prompts are often created using f-strings, which are not reusable.

Example (incorrect approach):
Using f-strings hardcodes the prompt and reduces flexibility.

To solve this, we use PromptTemplate from LangChain, which allows us to:
- Define reusable prompt structures
- Dynamically insert values
- Build scalable prompt systems

In this step, we convert a hardcoded prompt into a reusable PromptTemplate.

```python
# Task 1: Replace hardcoded prompt with PromptTemplate

# Define template
template = "Explain {topic} in simple terms for beginners"

# Create PromptTemplate
prompt = PromptTemplate(
    input_variables=["topic"],
    template=template
)

# Test the template
output = prompt.format(topic="Artificial Intelligence")

print("Generated Prompt:")
print(output)
```

# Task 2: Multi-Input Prompt System

In real-world applications, prompts are not built using a single input.
They often require multiple inputs such as topic, audience, and tone.

LangChain allows us to create dynamic prompts with multiple input variables.

In this step, we:
- Create a PromptTemplate with multiple inputs
- Generate prompts dynamically using different test cases

This improves flexibility and makes the system more reusable.

```python
# Task 2: Multi-Input PromptTemplate

# Define template
template = "Explain {topic} for {audience} in a {tone} tone"

# Create PromptTemplate
multi_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

# Test cases
print("Test Case 1:")
print(multi_prompt.format(topic="AI", audience="beginners", tone="friendly"))

print("\nTest Case 2:")
print(multi_prompt.format(topic="Python", audience="kids", tone="fun"))

print("\nTest Case 3:")
print(multi_prompt.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

# Task 3: Prompt Variations Engine

In real-world GenAI systems, we often generate different types of prompts
based on the same input (topic), depending on the use case.

For example:
- Teaching → Explain step-by-step
- Interview → Ask questions
- Storytelling → Explain as a story

In this step, we create multiple PromptTemplates for different styles
and generate outputs using the same topic.

This demonstrates how prompts can be customized for different scenarios.

```python
# Task 3: Prompt Variations

# Teaching template
teaching_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview template
interview_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

# Storytelling template
story_prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

# Input
topic = "Machine Learning"

print("Teaching Style:")
print(teaching_prompt.format(topic=topic))

print("\nInterview Style:")
print(interview_prompt.format(topic=topic))

print("\nStorytelling Style:")
print(story_prompt.format(topic=topic))
```

# Task 4: ChatPromptTemplate (Role-Based Prompting)

In modern LLM applications (like ChatGPT), prompts are structured as conversations
instead of plain text.

LangChain provides ChatPromptTemplate, which allows us to define:
- System role → Defines behavior (e.g., teacher, interviewer)
- User role → Provides input

This helps simulate real chat-based AI systems.

In this step, we:
- Create a chat-based prompt
- Use different roles like teacher, interviewer, and motivator
- Generate structured messages

```python
# Task 4: ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role}"),
    ("user", "Explain {topic}")
])

# Test with different roles
roles = ["teacher", "interviewer", "motivator"]

for role in roles:
    print(f"\nRole: {role}")

    messages = chat_template.format_messages(
        role=role,
        topic="Neural Networks"
    )

    for msg in messages:
        print(msg.content)
```

# Task 5: Input Validation Layer

In real-world applications, user inputs may be incorrect or invalid.

To make our prompt system robust, we implement an input validation layer.

Rules:
- audience must be one of: beginner, intermediate, expert
- tone must be one of: formal, casual, fun

If invalid input is provided:
- Assign default values
- OR handle errors

This ensures reliability and prevents incorrect prompt generation.

```python
# Task 5: Input Validation

def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Validate audience
    if audience not in valid_audience:
        print("Invalid audience → defaulting to 'beginner'")
        audience = "beginner"

    # Validate tone
    if tone not in valid_tone:
        print("Invalid tone → defaulting to 'formal'")
        tone = "formal"

    return audience, tone
```

```python
# Testing validation

print(validate_inputs("kids", "fun"))        # invalid audience
print(validate_inputs("expert", "serious"))  # invalid tone
print(validate_inputs("beginner", "casual")) # valid inputs
```

# Task 6: Prompt Generator Application

In this step, we build a complete prompt generation system.

This function integrates:
- Input validation
- Dynamic PromptTemplate creation
- Different styles (teaching, interview, storytelling)

The function takes:
- topic
- audience
- tone
- style

And generates a structured prompt dynamically.

This simulates a real-world GenAI pipeline:
User Input → Validation → Template → Output

```python
# Task 6: Prompt Generator Function

def generate_prompt(topic, audience, tone, style):

    # Step 1: Validate inputs
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Select template based on style
    if style == "teaching":
        template = "Explain {topic} for {audience} in a {tone} teaching style"

    elif style == "interview":
        template = "Generate 3 interview questions on {topic} for {audience} in a {tone} tone"

    elif style == "storytelling":
        template = "Explain {topic} for {audience} in a {tone} storytelling style"

    else:
        return "Invalid style"

    # Step 3: Create PromptTemplate
    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    # Step 4: Generate output
    return prompt.format(topic=topic, audience=audience, tone=tone)
```

```python
# Test the prompt generator

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
print(generate_prompt("Machine Learning", "expert", "formal", "interview"))
print(generate_prompt("Python", "kids", "fun", "teaching"))  # invalid audience test
```

# Task 7: Template Reusability Test

One of the main advantages of PromptTemplate is reusability.

Instead of creating multiple templates, we can use a single template
and generate different outputs by changing input values.

In this step, we:
- Use ONE PromptTemplate
- Pass multiple inputs
- Generate different outputs with the same structure

This demonstrates modular design and reusability in prompt engineering.

```python
# Task 7: Reusability Test

# Single reusable template
template = "Explain {topic} for {audience} in a {tone} tone"

reuse_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template
)

# Multiple inputs
test_cases = [
    ("AI", "beginner", "fun"),
    ("Python", "intermediate", "formal"),
    ("Machine Learning", "expert", "casual"),
    ("Deep Learning", "beginner", "formal"),
    ("NLP", "expert", "fun")
]

# Generate outputs
for topic, audience, tone in test_cases:
    print(reuse_prompt.format(topic=topic, audience=audience, tone=tone))
```
